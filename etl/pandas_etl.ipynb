```python
import pandas as pd
from sqlalchemy import create_engine
```

### Read in csv file and transform to load to tables

```python
first_file = "../NPS Data/Annual Park Recreation Visitation (1904 - Last Calendar Year).csv"
first_df = pd.read_csv(first_file)
first_df.head()
```

```python
first_df.count
```

```python
first_cols = ["Year", "RecreationVisitors"]
first_transformed= first_df[first_cols].copy()

# Rename the column headers
first_transformed = first_transformed.rename(columns={"Year": "year",
                                                     "RecreationVisitors": "recreationvisitors"})
                                                           

# Clean the data by dropping duplicates and setting the index
first_transformed.drop_duplicates("year", inplace=True)
first_transformed.set_index("year", inplace=True)

first_transformed.head()
```

```python
 
#first_transformed["test"] = first_transformed["year"].astype("datetime64[ns]")
#first_transformed["test"]
#first_transformed["recreationvisitors"].astype(str).astype(int)
first_transformed.replace(',','', regex=True, inplace=True)
```

```python
first_transformed.count
```

```python
second_file = "../NPS Data/Overnight Stays (1979 - Last Calendar Year).csv"
second_df = pd.read_csv(second_file)
second_df.head()
```

```python
second_df.count
```

```python
second_cols = ["Year",
               "Concessioner Lodgers", 
               "Concessioner Campers",
               "Tent Campers", 
               "RV Campers", 
               "Backcountry Campers",
               "Miscellaneous Campers"
              ]
second_transformed = second_df[second_cols].copy()

# Rename the column headers
second_transformed = second_transformed.rename(columns={"Year": "year",
                                                        "Concessioner Lodgers": "concessionerlodgers",
                                                        "Concessioner Campers": "concessionercampers",
                                                        "Tent Campers": "tentcampers",
                                                        "RV Campers": "rvcampers",
                                                        "Backcountry Campers": "backcountrycampers",
                                                        "Miscellaneous Campers": "miscellaneouscampers"})
                                                       
                                                           

# Clean the data by dropping duplicates and setting the index
second_transformed.drop_duplicates("year", inplace=True)
second_transformed.set_index("year", inplace=True)

second_transformed.head()
```

```python
second_transformed.replace(',','', regex=True, inplace=True)
```

```python
second_transformed.count
```

```python
third_file = "../NPS Data/Summary of Visitor Use By Month and Year (1979 - Last Calendar Year).csv"
third_df = pd.read_csv(third_file)
third_df.head()
```

```python
third_df.count
```

```python
third_cols = ["Year",
               "Month", 
               "Receation Visitiors",
               "Non-Recreation Visitors", 
               "RecreationHours", 
               "NonRecreationHours",
               "Tent Campers",
               "RV Campers",
               "TotalOvernightStays" 
              ]
third_transformed = third_df[third_cols].copy()

# Rename the column headers
third_transformed = third_transformed.rename(columns={"Year": "year",
                                                      "Month": "month",
                                                      "Receation Visitiors": "recreationvisitors",
                                                      "Non-Recreation Visitors": "nonrecreationvisitors",
                                                      "RecreationHours": "recreationhours",
                                                      "NonRecreationHours": "nonrecreationhours",
                                                      "Tent Campers": "tentcampers",
                                                      "RV Campers": "rvcampers",
                                                      "TotalOvernightStays": "totalovernightstays"})
                                                       
                                                           

# Clean the data by dropping duplicates and setting the index
third_transformed.drop_duplicates(inplace=True)
#third_transformed.set_index("year", inplace=True)

third_transformed.head()
```

```python
third_transformed.replace(',','', regex=True, inplace=True)
```

```python
third_transformed.count
```

### Create database connection

```python
# Connect to the database
from config import protocol, username, password, host, port, database_name

rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

```python
# Confirm tables
engine.table_names()
```

### Load DataFrames into database

```python
first_transformed.to_sql(name='yr_visitors', con=engine, if_exists='append', index=True)
```

```python
second_transformed.to_sql(name='lodge_ovn', con=engine, if_exists='append', index=True)
```

```python
third_transformed.to_sql(name= 'sum_visitors', con=engine, if_exists='append', index=False)
```
